```python
import pandas as pd
from funpymodeling.exploratory import freq_tbl, status
```

```python
# Para este caso nos interesa visualizar todas las columnas
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("water_potability.csv", sep=',') 
```

```python
data.head(3)
```

```python
status(data)
```

```python
# Pasamos a discretizar

data['ph_cat'], saved_bins_1 = pd.qcut(data['ph'],
                              q=5,
                              retbins=True) # importante!

data['Sulfate_cat'], saved_bins_2 = pd.qcut(data['Sulfate'],
                              q=5,
                              retbins=True) # importante!

data['Trihalomethanes_cat'], saved_bins_3 = pd.qcut(data['Trihalomethanes'],
                              q=5,
                              retbins=True) # importante!
```

```python
# (Podría borrar o no las columnas que he discretizado y dejo las discretizados, depende del caso) En este caso las borro. Dejo las _cat ..

data = data.drop(['ph'], axis=1)
data = data.drop(['Sulfate'], axis=1)
data = data.drop(['Trihalomethanes'], axis=1)
```

```python
status(data)
#freq_tbl(data['Sulfate_cat'])
```

```python
# Primero creo la categoría "desconocido" y luego le asigno con FILLNA ese valor categórico a los Nan de la columnas que he discretizado en el paso anterior.

data['ph_cat']=data['ph_cat'].cat.add_categories("desconocido")
data['ph_cat']=data['ph_cat'].fillna(value="desconocido")

data['Sulfate_cat']=data['Sulfate_cat'].cat.add_categories("desconocido")
data['Sulfate_cat']=data['Sulfate_cat'].fillna(value="desconocido")

data['Trihalomethanes_cat']=data['Trihalomethanes_cat'].cat.add_categories("desconocido")
data['Trihalomethanes_cat']=data['Trihalomethanes_cat'].fillna(value="desconocido")
```

```python
status(data)
#freq_tbl(data['Sulfate_cat'])
```

```python
data_ohe = pd.get_dummies(data)
```

```python
status(data_ohe)
```

```python
data_x = data_ohe.drop('Potability', axis=1)
data_y = data_ohe['Potability']
```

```python
status(data_y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier 

# Creamos 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 99)
```

```python
rf.fit(x_train, y_train)
```

```python
import pickle

# Guardar en el disco
filename = 'rf.pkl'
pickle.dump(rf, open(filename, 'wb')) # rf = nuestro modelo

# Lo cargamos para usarlo en otro momento. Asi:
# rf_loaded = pickle.load(open(filename, 'rb'))

# Guardamos las columnas x (sin Potability)
with open('categories_ohe.pickle', 'wb') as handle:
    pickle.dump(data_x.columns, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
